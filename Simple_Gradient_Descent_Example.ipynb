```python
#Import packages. Feel free to add any others you might use!
import itertools
import jax.numpy as jnp
import jax
import matplotlib.pyplot as plt
from jax import random as jrandom
from jax import nn as jnn
import random
import time
import math
import numpy as np
```

```python
#Suppose we have some inputs X
X = jnp.array([1,2,3],dtype=float)
#And an equal number of outputs Y 
Y = jnp.array([4,8,10],dtype=float)
assert jnp.shape(X) == jnp.shape(Y)
plt.plot(X,Y,linestyle="",marker="o")
```

```python
#We would like to find a model function f(x;w)
# which depends on some weights w
# and which hopefully helps us explain our data
def f(x,w):
  return w[0]+w[1]*x 

#Here is what the graph of f looks like
def plot_f(w):
  x_values = jnp.arange(-3,3,0.1)
  fx_values = f(x_values,w)
  plt.plot(x_values,fx_values)
  return True

theta = jnp.array([0.5, -0.5])
plot_f(theta)
```

```python
#The "loss" measures how well our function fits the data 
# i.e. how close f(X,w) is to Y
def loss(w):
  return jnp.mean(jnp.square(f(X,w) - Y))

#The gradient of the loss is a new function that tells us the derivatve of the loss
# w.r.t. all of the weights w
grad_loss = jax.grad(loss)
```

```python
#Here is a test of the gradient
w = jnp.array([0.5, -0.5])
grad_loss(w)
```

```python
w = jnp.array([0.0,0.0]) #initial weights 
alpha = 0.01 #learning rate
num_iterations = 100

#Gradient descent algorithm!
for i in range(num_iterations):
  w -= alpha*grad_loss(w) #Gradient descent!
  plt.plot(X,Y,linestyle="",marker="o")
  plot_f(w)

print(w)
```
