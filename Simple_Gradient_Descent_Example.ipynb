```python
#Import packages. Feel free to add any others you might use!
import itertools
import jax.numpy as jnp
import jax
import matplotlib.pyplot as plt
from jax import random as jrandom
from jax import nn as jnn
import random
import time
import math
import numpy as np
```

```python
X = jnp.array([1,2,3],dtype=float)
Y = jnp.array([4,8,10],dtype=float)
assert jnp.shape(X) == jnp.shape(Y)
plt.plot(X,Y,linestyle="",marker="o")
```

```python
def f(x,theta):
  return theta*x 

def plot_f(theta):
  x_values = jnp.arange(-3,3,0.1)
  fx_values = f(x_values,theta)
  plt.plot(x_values,fx_values)
  return True

plot_f(0.5)
```

```python
def loss(theta):
  return jnp.mean(jnp.square(f(X,theta) - Y))

grad_loss = jax.grad(loss)
```

```python
grad_loss(0.0)
```

```python
theta = 0.0 #initial parameter (make sure its a float!)
alpha = 0.01 #learning rate
num_iterations = 100
for i in range(num_iterations):
  theta -= alpha*grad_loss(theta) #Gradient descent!
  plt.plot(X,Y,linestyle="",marker="o")
  plot_f(theta)
```
